```python
import numpy as np
import pandas as pd
import sqlite3 as sql
from IPython.core.display import Image
```

```python
# connect to SQLite database
db_path = "../../Data/basketball.sqlite"
connection = sql.connect(db_path) # create connection object to database
table = pd.read_sql_query("SELECT name FROM sqlite_master WHERE type='table'", connection)
print(table)
```

```python
#Number of games
gamesQ = """SELECT * FROM 'Game'"""

game_df = pd.read_sql(gamesQ, connection)
game_df.head()
```

```python
game_df.to_csv("../DataTables/games.csv",index=False)
```

```python
Players_attrib_Q = """SELECT * FROM 'Player_Attributes'"""

Players_attrib_df = pd.read_sql(Players_attrib_Q, connection)
Players_attrib_df.head()
```

```python
Players_attrib_df.to_csv("../DataTables/PlayersAttributes.csv",index=False)
```

```python
PlayersQ = """SELECT * FROM 'Player'"""

Players_df = pd.read_sql(PlayersQ, connection)
Players_df.to_csv("../DataTables/Players.csv",index=False)
```
